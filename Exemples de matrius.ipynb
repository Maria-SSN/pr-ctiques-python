Valors i vectors propis

```python
A=np.array([[2,2],[1,3]])
```

```python
Valors, vectors=np.linalg.eig(A)
print("valors propis",Valors)
print("vectors propis", vectors)
```

```python
u=vectors[:,0]
print (u)
```

```python
v=vectors[:,1]
print(v)
```

Comprovem que son vectors propis

```python
z=np.dot(A,v)
print (z)
```

```python
z2=np.dot(A,u)
print(z2)
```
